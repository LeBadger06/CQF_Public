### Introduction

#### Business setting

Business entities such as oil companies or banks may face the need to use, hedge and/or price complex derivative instruments such as Asian options, binary options or lookback options (we will focus on call options here). 
- Asian options give the holder a payoff that depends on the average price of the underlying over some prescribed period. This typically reduces the option premium, as the volatility of the average is lower than that of the underlying. It also reduces exposure to sudden movements in the underlying before expiry. Such attributes render them widely attractive to e.g. oil market participants.
- Binary options give the holder a payoff of \$1 if the underlying rises (or falls, for puts) above the strike at expiry. They have a much more limited payoff potential than their vanilla counterparts and express a more conservative view of the market.
- Lookback options give the holder a payoff that is some function of the realized maximum and/or minimum of the underlying over some prescribed period. The premium for these could be high since their payoff is also potentially high.

Some exact formulae may exist to price  some of these options; otherwise we need to resort to numerical methods such as finite differences or Monte Carlo (MC) simulation. In this assignment, we will focus on exact formulae and MC simulation. 
- Advantages of MC simulation include: relative ease-of-use, ability to include path-dependency;
- Disadvantages of MC simulation include: speed of convergence, potential inaccuracy.

#### Mathematical setting

Let $S(t)$ be the price of the underlying, assumed to follow a risk-neutral random walk
$$
dS = rSdt + \sigma SdX
$$
where the drift $\mu$ has been replaced with the interest rate $r$, and $\sigma$ is constant volatility and $X$ is the Brownian motion. That is: $S$ follows a Geometric Brownian Motion (GBM).
Let $V(S,t)$ be the value of an option with a certain payoff function $\text{Payoff}$ and expiry $T$. Theory shows that $V$ is the expected value of the discounted payoff under the risk-neutral density $Q$:
$$
V(S,t) = \exp(-r(T-t))E^Q[\text{Payoff}].
$$

#### Euler-Maruyama (EM) scheme

The forward EM scheme for GBM is: 
$$
S_{t+\delta t} \sim S_{t}\left(1+r\delta t+\sigma\phi\sqrt{\delta t}\right)
$$
where $\phi \sim N(0,1)$ is the same random number used per time step. The approximation error in EM is $O(\sqrt{\delta t})$.

#### Milstein scheme

The Milstein scheme for GBM is: 
$$
S_{t+\delta t} \sim S_{t}\left(1+r\delta t+\sigma\phi\sqrt{\delta t}+\frac{1}{2}\sigma^2(\phi^2 - 1)\delta t\right)
$$
where $\phi \sim N(0,1)$ is the same random number used per time step. The approximation error in Milstein is $O(\delta t)$, better than EM.

#### Options under consideration
- Asian call options: the payoff is given by:
$$
\text{Payoff} = C_{T} = max(0,\frac{1}{T}\sum_{i=1}^{T}{S_i-K}) = \exp(-r(T-t))E^{Q}[max(0,S_{avg}-K)]
$$ under the risk-neutral framework.
- Binary call options give a \$1 payoff if the underlying exceeds the strike at expiry.
- Lookback options: a floating-strike lookback call gives the holder the right to buy the underlying at the lowest price observed $S_{min}$ during the option lifetime; the payoff is:
$$
C(S,S_{min},T)=max(S-S_{min},0).
$$ 
There are more sophisticated types of Asian and lookback options but we will focus on these only here.

#### Stochastic volatility

Assuming stochastic volatility (Heston model, described in PWoQF pp. 860-861), the mathematical setting becomes:
$$
dS = rSdt + \sigma SdX
$$
$$
d\sigma^2 = \kappa(\theta - \sigma^2)dt+\chi\sigma dW
$$
where $X$ and $W$ are correlated Brownian motions, $\kappa$ is the rate of mean-reversion, $\theta$ is the long-run average variance and $\chi$ is the volatility of volatility. In that setting, the EM scheme becomes:
$$
S_{t+\delta t} \sim S_{t}(1+r\delta t+\sigma_t\phi\sqrt{\delta t})
$$
$$
\sigma^2_{t+\delta t} \sim \sigma^2_{t} + \kappa(\theta - \sigma^2_t)dt+\chi\sigma_t \gamma\sqrt{\delta t}
$$
where $\phi$ and $\gamma$ are correlated. 

We will do most of our work assuming constant volatility, but will also test the EM scheme and perform some sensitivity analysis assuming stochastic volatility in the last section, inspired by chapter 18 of (Hilpisch, 2019). The Milstein scheme with stochastic volatility is somewhat more complicated and we do not attempt it here. See chapter 4 of (Duffy & Kienitz, 2009) for more details.

### Closed form solutions

Let $N$ denote the standard normal c.d.f.

#### Closed form solution with Asian call option

A closed form expression for geometric averaging Asian call options (in the below we assume no dividends) can be found in (Haug, 2007; p. 183). Let:
$$
b_A = \frac{1}{2}(r-\frac{\sigma^2}{6}),
$$
$$
\sigma_A=\frac{\sigma}{\sqrt{3}},
$$
$$
d_1=\frac{ln(S_0/K)+(b_A+\frac{\sigma^2_A}{2})(T-t)}{\sigma_A\sqrt{(T-t)}},
$$
$$
d_2 = d_1-\sigma_a\sqrt{(T-t)},\text{ then}
$$
$$
C_T = S_0 \exp{((b_A-r)(T-t))}N(d_1)-K\exp{(-r(T-t))}N(d_2).
$$



```python
def closed_form_solution_Asian(spot, strike, rate, dte, volatility):
    b_A = 0.5*(rate- volatility*volatility/6)
    vol_avg = volatility/math.sqrt(3)
    f = e**((b_A-r)*dte)
    g = e**(-r*dte)
    d_1 = (log(spot / strike) +  (b_A + 0.5*vol_avg*vol_avg)*dte) / (vol_avg* dte**0.5)
    d_2 = d_1 - vol_avg* dte**0.5 
    
    value = spot * f * norm.cdf(d_1) - strike * g * norm.cdf(d_2)
    
    return value 
```

#### Closed form solution with binary call option

The binary call option value is (from PWoQF p. 120):
$$
C_T=\exp(-r(T-t))N(d_2).
$$

```python
# define closed form solution functions
def closed_form_solution(spot, strike, rate, dte, volatility, type):
    
    a = volatility * dte**0.5
    d_1 = (log(spot / strike) +  (rate + (volatility**2) / 2) * dte) / a 
    d_2 = d_1 - a
    b = e**-(rate * dte)
    
    if type == 1: # vanilla call 
        value = spot * norm.cdf(d_1) - strike * b * norm.cdf(d_2)
    elif type == 2: # binary
        value = b * norm.cdf(d_2) # from PWoQF, p. 120
    
    return value 
```

#### Closed form solution with lookback call option

Let $b$ be the cost of carry (for an option on stocks, $b= r$); we have (from Haug, 2007; p. 142):
$$
a_1=\frac{ln(S/S_{min})+(b+\sigma^2/2)(T-t)}{\sigma\sqrt{(T-t)}}
$$
$$
a_2=a_1-\sigma\sqrt{(T-t)}
$$
$$
f=\left[\left(\frac{S}{S_{min}}\right)^{-\frac{2b}{\sigma^2}}N(-a_1+\frac{2b}{\sigma}\sqrt{(T-t)})-\exp{(b(T-t))N(-a_1)}\right]
$$
$$
C_T=S_0\exp{((b-r)(T-t))}N(a_1)-S_{min}\exp(-r(T-t))N(a_2)+S_0\exp{(-r(T-t))}\frac{\sigma^2}{2b}f
$$.

```python
def closed_form_solution_lookback(spot, s_min, rate, dte, volatility):
    a_1 = (log(spot/s_min) + (rate+0.5*volatility*volatility)*dte) / (volatility * dte**0.5)
    a_2 = a_1 - volatility*(dte**0.5)
    f_1 = (spot/s_min)**(-2*rate/(volatility*volatility))*norm.cdf(-a_1+2*rate*(dte**0.5)/volatility)
    f_2 = e**(rate*dte)*norm.cdf(-a_1)
    f = f_1-f_2
    g = volatility*volatility/(2*rate)
    
    value = spot * norm.cdf(a_1) - s_min * (e**(-rate*dte)) * norm.cdf(a_2) + spot * (e**(-rate*dte))*g*f
    
    return value 
```

### EM Scheme

In this section, we implement the EM scheme for the three types of call options under consideration and get a feel for for their values for ATM options (assuming 252 trading days in a year).

#### EM Scheme with Asian call option

```python
def simulate_path_EM(s0, mu, sigma, horizon, timesteps, n_sims):

    # set the random seed for reproducibility
    random.seed(10000)

    # read parameters
    S0 = s0                 # initial spot price
    r = mu                  # mu = rf in risk neutral framework
    T = horizon             # time horizon
    t = timesteps           # number of time steps
    n = n_sims              # number of simulation

    # define dt
    dt = T/t                # length of time interval

    # simulate 'n' asset price path with 't' timesteps
    S = zeros((t,n))
    S[0] = S0

    for i in range(0, t-1):
        w = random.standard_normal(n)
        S[i+1] = S[i] * ( 1 + r*dt + sigma*sqrt(dt)*w )

    return S
```

```python
S = simulate_path_EM(S_0,r,sigma,T,trading_days_in_yr,num_sims)

# Average price
A = S.mean(axis=0)
    
C0 = exp(-r*T) * mean(maximum(0, A-K))

print(f"Asian Call Option Value is {C0:0.10f}")
```

##### EM Scheme w Asian call option - antithetic

We implement here the EM scheme with antithetic variables for variance reduction and illustrate it with Asian options. 

```python
def simulate_path_EM_anti(s0, mu, sigma, horizon, timesteps, n_sims):

    # set the random seed for reproducibility
    random.seed(10000)

    # read parameters
    S0 = s0                 # initial spot price
    r = mu                  # mu = rf in risk neutral framework
    T = horizon             # time horizon
    t = timesteps           # number of time steps
    n = n_sims              # number of simulation

    # define dt
    dt = T/t                # length of time interval

    # simulate 'n' asset price path with 't' timesteps
    S = zeros((t,n))
    S[0] = S0
    U = zeros((t,n))
    U[0] = S0
    

    for i in range(0, t-1):
        w = random.standard_normal(n)
        S[i+1] = S[i] * (1 + r*dt + sigma*sqrt(dt)*w)
        U[i+1] = U[i] * (1 + r*dt + sigma*sqrt(dt)*(-w)) # note the minus sign

    return S, U
```

```python
S, U = simulate_path_EM_anti(S_0,r,sigma,T,trading_days_in_yr,num_sims)

# Average price
A = S.mean(axis=0)
B = U.mean(axis=0)

C0_anti = 0.5*exp(-r*T) * mean(maximum(0, A-K)+maximum(0, B-K))

print(f"Asian Call Option Value is {C0_anti:0.10f}")
```

#### EM Scheme with binary call option

```python
S = simulate_path_EM(S_0,r,sigma,T,trading_days_in_yr,num_sims)

# There must be a better way, but it works
run_sum = 0
for i in range(num_sims):
    if S[-1,i] >= K:
      run_sum = run_sum+1  
avg_sum = run_sum/num_sims

C0 = exp(-r*T) * avg_sum

print(f"Binary Call Option Value is {C0:0.5f}")
```

#### EM Scheme with lookback option

```python
S = simulate_path_EM(S_0,r,sigma,T,trading_days_in_yr,num_sims)

S_min = S.min(axis = 0)

C0 = exp(-r*T) * mean(maximum(0, S[-1]-S_min))

print(f"Lookback Call Option Value is {C0:0.5f}")
```

Observation: The lookback option appears a lot more expensive than the Asian option, which itself is a lot more expensive than the binary option. 

### Milstein Scheme

In this section, we implement the Milstein scheme for the three types of call options under consideration and get a feel for for their values for ATM options (assuming 252 trading days in a year).

#### Milstein scheme with Asian call option

```python
def simulate_path_Milstein(s0, mu, sigma, horizon, timesteps, n_sims):

    # set the random seed for reproducibility
    random.seed(10000)

    # read parameters
    S0 = s0                 # initial spot price
    r = mu                  # mu = rf in risk neutral framework
    T = horizon             # time horizon
    t = timesteps           # number of time steps
    n = n_sims              # number of simulation

    # define dt
    dt = T/t                # length of time interval

    # simulate 'n' asset price path with 't' timesteps
    S = zeros((t,n))
    S[0] = S0

    for i in range(0, t-1):
        w = random.standard_normal(n)
        S[i+1] = S[i] * ( 1 + r*dt + sigma*sqrt(dt)*w + 0.5*sigma*sigma*(w*w-1)*dt )
    return S
```

### Speed of convergence: EM vs. Milstein for GBM

We introduce a function that simulates both schemes in one go:

```python
def simulate_paths_EM_Milstein(s0, mu, sigma, horizon, timesteps, n_sims):

    # set the random seed for reproducibility
    random.seed(10000)

    # read parameters
    S0 = s0                 # initial spot price
    r = mu                  # mu = rf in risk neutral framework
    T = horizon             # time horizon
    t = timesteps           # number of time steps
    n = n_sims              # number of simulation

    # define dt
    dt = T/t                # length of time interval

    # simulate 'n' asset price path with 't' timesteps
    S_EM = zeros((t,n))
    S_EM[0] = S0
    S_Mil = zeros((t,n))
    S_Mil[0] = S0
    
    sch = dict()

    for i in range(0, t-1):
        w = random.standard_normal(n)
        S_EM[i+1] = S_EM[i] * ( 1 + r*dt + sigma*sqrt(dt)*w  )
        S_Mil[i+1] = S_Mil[i] * ( 1 + r*dt + sigma*sqrt(dt)*w + 0.5*sigma*sigma*(w*w-1)*dt )
        # note that we use the same w in both schemes for the same timestep i+1
        
    sch['EM'] = S_EM
    sch['Mil'] = S_Mil
    
    return sch
```

We will increase the number of timesteps in the EM and Milstein schemes for the GBM and observe whether the error vs. the exact solution decreases. 

```python
# Attempt at linspace
tstep_range = np.linspace(2,100, 20, endpoint = True)
tstep_range
```

```python
def closed_form_GBM_mean(S_0, r, T):  
    return S_0*exp(r*T)

def EM_anti(S_0,K,r,sigma,T,timesteps,num_sims):
    S, U = simulate_path_EM_anti(S_0,r,sigma,T,timesteps,num_sims)
    value_A = mean(S[-1])
    value_B = mean(U[-1])
    value = 0.5*(value_A+value_B)
    return value
```

```python
def compare_EM_Milstein(S_0,K,r,sigma,T,tstep_range, num_sims): 
    res = arange(1,5,1)
    MCconv = zeros((len(tstep_range),len(res)))
    MCconv = pd.DataFrame(MCconv, index = tstep_range, columns=['Exact', 'EM', 'Milstein', 'EM_anti'])
    MCconv.index = MCconv.index.astype(int)
    for i in range(0, len(tstep_range)):
        sch = simulate_paths_EM_Milstein(S_0,r,sigma,T,int(T*tstep_range[i]),num_sims)
        for k in range(0, len(res)):
            if k == 0:
                MCconv.iloc[i,k] = closed_form_GBM_mean(S_0, r, T)
            elif k == 1:
                MCconv.iloc[i,k] = mean(sch['EM'][-1])
            elif k == 2:
                MCconv.iloc[i,k] = mean(sch['Mil'][-1])
            elif k == 3:
                MCconv.iloc[i,k] = EM_anti(S_0,K,r,sigma,T,int(T*tstep_range[i]),num_sims)
    MCconv['EM_err_perc'] = 100*(MCconv['EM']-MCconv['Exact'])/MCconv['Exact']
    MCconv['Milstein_err_perc'] = 100*(MCconv['Milstein']-MCconv['Exact'])/MCconv['Exact']
    MCconv['EM_anti_err_perc'] = 100*(MCconv['EM_anti']-MCconv['Exact'])/MCconv['Exact']
    return MCconv
```

### EM Scheme with stochastic volatility

We implement the EM scheme with stochastic volatility and test it for various call options below. The Milstein scheme with stochatic volatility is somewhat more complex as described in chapter 4 of Duffy & Kienitz and we leave it for future research.

```python
import numpy as np
def simulate_path_Heston(S_0, K, r, T, sigma, kappa, theta, volvol, rho, time_steps_per_yr, num_sims):
    timesteps = T * time_steps_per_yr
    dt = 1/time_steps_per_yr
    
    S_t = np.zeros((timesteps, num_sims))
    V_t = np.zeros((timesteps, num_sims))
    
    # Initialize
    V_t[0,:] = sigma
    S_t[0, :] = S_0
    
    # Use Cholesky decomposition
    means = [0,0]
    stdevs = [1/3, 1/3]
    covs = [[stdevs[0]**2, stdevs[0]*stdevs[1]*rho], 
            [stdevs[0]*stdevs[1]*rho,stdevs[1]**2]]

    Z = np.random.multivariate_normal(means, covs, (num_sims, timesteps)).T
    Z1 = Z[0]
    Z2 = Z[1]
    for i in range(1, timesteps):
        # Use Z2 to calculate Vt
        V_t[i,:] = np.maximum(V_t[i-1,:] + kappa * (theta - V_t[i-1,:])* dt + volvol *  np.sqrt(V_t[i-1,:] * dt) * Z2[i,:],0)
        
        # Use all V_t calculated to find the value of S_t
        S_t[i,:] = S_t[i-1,:] + r * S_t[i,:] * dt + np.sqrt(V_t[i,:] * dt) * S_t[i-1,:] * Z1[i,:]
    return S_t
```

```python
# Parameters of the Heston model
# These would need to be calibrated to empirical data
rho = 0.25; # correlation between underlying and its volatility, assumed fairly low here
volvol = 0.15; # volatility of volatility 
theta = 0.1; # long-run avg variance of the underlying
kappa = 0.2; # rate at which sigma^2 reverts to theta
```

```python
def Asian_EM_heston(S_0,K,r,sigma,T,timesteps,num_sims):
    S = simulate_path_Heston(S_0, K, r, T, sigma, kappa, theta, volvol, rho, timesteps, num_sims)
    A = S.mean(axis=0)
    value = exp(-r*T) * mean(maximum(0, A-K))
    return value 
    
C0 = Asian_EM_heston(S_0,K,r,sigma,T,trading_days_in_yr,num_sims)

# Print the values
print(f"Asian Call Option Value is (Heston model) {C0:0.10f}")
```

As volatility becomes stochastic, we notice already a drop in the Asian call value compared to the constant volatility case.

Next, we vary the number of runs in the EM scheme and compare the EM scheme with constant volatility and the EM scheme with stochastic volatility and observe the effect on the option value.

```python
def run_methods_Asian(S_0,K,r,sigma,T,timesteps):
    simrange = arange(min_sim, max_sim, sim_step)
    res = arange(1,6,1)
    MCconv = zeros((len(simrange),len(res)))
    MCconv = pd.DataFrame(MCconv, index = simrange, columns=['Exact', 'EM', 'Milstein', 'EM_anti', 'EM_Heston'])
    for i in range(0, len(simrange)):
        for k in range(0, len(res)):
            sch = simulate_paths_EM_Milstein(S_0,r,sigma,T,timesteps,simrange[i])
            if k == 0:
                MCconv.iloc[i,k] = closed_form_solution_Asian(S_0, K, r, T, sigma)
            elif k == 1:
                MCconv.iloc[i,k] = Asian_Call_Payoff(sch['EM'],r,T,K)
            elif k == 2:
                MCconv.iloc[i,k] = Asian_Call_Payoff(sch['Mil'],r,T,K)
            elif k == 3:
                MCconv.iloc[i,k] = Asian_EM_anti(S_0,K,r,sigma,T,timesteps,simrange[i])
            elif k == 4:
                MCconv.iloc[i,k] = Asian_EM_heston(S_0,K,r,sigma,T,timesteps,simrange[i])
    MCconv['EM_err_perc'] = 100*(MCconv['EM']-MCconv['Exact'])/MCconv['Exact']
    MCconv['Milstein_err_perc'] = 100*(MCconv['Milstein']-MCconv['Exact'])/MCconv['Exact']
    MCconv['EM_anti_err_perc'] = 100*(MCconv['EM_anti']-MCconv['Exact'])/MCconv['Exact']
    MCconv['EM_Heston_err_perc'] = 100*(MCconv['EM_Heston']-MCconv['Exact'])/MCconv['Exact']
    return MCconv
```

#### Binary call option

```python
def Binary_EM_heston(S_0,K,r,sigma,T,timesteps,num_sims):
    S = simulate_path_Heston(S_0, K, r, T, sigma, kappa, theta, volvol, rho, timesteps, num_sims)
    run_sum = 0
    for i in range(num_sims):
            if S[-1,i] >= K:
              run_sum = run_sum+1  
    avg_sum = run_sum/num_sims
    value = exp(-r*T) * avg_sum 
    return value 
```

```python
def run_methods_Binary(S_0,K,r,sigma,T,timesteps):
    simrange = arange(min_sim, max_sim, sim_step)
    res = arange(1,6,1)
    MCconv = zeros((len(simrange),len(res)))
    MCconv = pd.DataFrame(MCconv, index = simrange, columns=['Exact', 'EM', 'Milstein', 'EM_anti', 'EM_Heston'])
    for i in range(0, len(simrange)):
        sch = simulate_paths_EM_Milstein(S_0,r,sigma,T,timesteps,simrange[i])
        for k in range(0, len(res)):
            if k == 0:
                MCconv.iloc[i,k] = closed_form_solution(S_0, K, r, T, sigma, 2)
            elif k == 1:
                MCconv.iloc[i,k] = Binary_Call_Payoff(sch['EM'],r,T,K)
            elif k == 2:
                MCconv.iloc[i,k] = Binary_Call_Payoff(sch['Mil'],r,T,K)
            elif k == 3:
                MCconv.iloc[i,k] = Binary_EM_anti(S_0,K,r,sigma,T,timesteps,simrange[i])
            elif k == 4:
                MCconv.iloc[i,k] = Binary_EM_heston(S_0,K,r,sigma,T,timesteps,simrange[i])
    MCconv['EM_err_perc'] = 100*(MCconv['EM']-MCconv['Exact'])/MCconv['Exact']
    MCconv['Milstein_err_perc'] = 100*(MCconv['Milstein']-MCconv['Exact'])/MCconv['Exact']
    MCconv['EM_anti_err_perc'] = 100*(MCconv['EM_anti']-MCconv['Exact'])/MCconv['Exact']
    MCconv['EM_Heston_err_perc'] = 100*(MCconv['EM_Heston']-MCconv['Exact'])/MCconv['Exact']
    return MCconv
```

#### Lookback call option

```python
def Lookback_EM_heston(S_0,K,r,sigma,T,timesteps,num_sims):
    S = simulate_path_Heston(S_0, K, r, T, sigma, kappa, theta, volvol, rho, timesteps, num_sims)
    S_min = S.min(axis = 0)
    value = exp(-r*T) * mean(maximum(0, S[-1]-S_min))
    return value 
```

```python
def run_methods_Lookback(S_0,K,r,sigma,T,timesteps):
    simrange = arange(min_sim, max_sim, sim_step)
    res = arange(1,6,1)
    MCconv = zeros((len(simrange),len(res)))
    MCconv = pd.DataFrame(MCconv, index = simrange, columns=['Exact', 'EM', 'Milstein', 'EM_anti', 'EM_Heston'])
    for i in range(0, len(simrange)):
        sch = simulate_paths_EM_Milstein(S_0,r,sigma,T,timesteps,simrange[i])
        for k in range(0, len(res)):
            if k == 0:
                MCconv.iloc[i,k] = Lookback_EM_exact(sch['EM'],r,T,S_0,sigma)
            elif k == 1:
                MCconv.iloc[i,k] = Lookback_Call_Payoff(sch['EM'],r,T)
            elif k == 2:
                MCconv.iloc[i,k] = Lookback_Call_Payoff(sch['Mil'],r,T)
            elif k == 3:
                MCconv.iloc[i,k] = Lookback_EM_anti(S_0,K,r,sigma,T,timesteps,simrange[i])
            elif k == 4:
                MCconv.iloc[i,k] = Lookback_EM_heston(S_0,K,r,sigma,T,timesteps,simrange[i])
    MCconv['EM_err_perc'] = 100*(MCconv['EM']-MCconv['Exact'])/MCconv['Exact']
    MCconv['Milstein_err_perc'] = 100*(MCconv['Milstein']-MCconv['Exact'])/MCconv['Exact']
    MCconv['EM_anti_err_perc'] = 100*(MCconv['EM_anti']-MCconv['Exact'])/MCconv['Exact']
    MCconv['EM_Heston_err_perc'] = 100*(MCconv['EM_Heston']-MCconv['Exact'])/MCconv['Exact']
    return MCconv
```

### Conclusion

Our main observations are:
- For GBM the error between EM and Milstein converges to zero for both schemes; however we do not observe a marked speedup in convergence for Milstein over EM in this particular setting.
- For Asian call options, the Milstein scheme provides smaller errors than the EM scheme for ATM and ITM options, but not so for OTM options, although the difference is small.
- For binary call options, the Milstein scheme provides smaller errors than the EM scheme for ITM options, but not so for OTM options or ATM, although the difference is small.
- For lookback call options, both the EM and Milstein schemes provide large errors ATM and ITM options, and the error diminishes for OTM options, but remains large still.
- The Asian call option is most sensitive to volatility near the strike. If it is far OTM or ITM, the sensitivity to volatility becomes small.
- As the interest rate and time to expiry increase, the Asian call option value increase as well. The impact of high interest rate and time to expiry give a higher option value than high volatility or starting point S_0.
- As the volatility increases the binary call option values decreases, which is somewhat counterintuitive.
- The binary call option appears a lot less sensitive to interest rates or time to expiry than other option types.
- The lookback call option value is a lot more sensitive to interest rate and time to expiry than to volatility and $S_0$.
- The exact formulae for Asian, binary or lookback call options assume constant volatility and may overestimate greatly the value of the option compared to a (more realistic) model with stochastic volality under the EM scheme.
